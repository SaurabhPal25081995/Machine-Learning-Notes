```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([2,4,6,4]) # here by default its treated as y axis value, and x axis value will be (0,1,2,3)
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title("My First Plot")
plt.show()
```

```python
# Grid

plt.plot([1,2,3,4],[1,4,9,16])  # here first array is for x-axis and 2nd one for y-axis
plt.ylabel("Squares")
plt.xlabel("Numbers")
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],"ro")  # here r means red and o means round
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(0., 5., 0.2)

plt.plot(t, t**2, 'b--', label = '^2')     # b means blue line and -- means dash line
plt.plot(t, t**2.2, 'rs', label = '^2.2')  # r means red line and s means square
plt.plot(t, t**2.5, 'g^', label = '^2.5')  # g means green and ^ means triangle
plt.grid()
plt.legend()
plt.show()
```

```python
# Controlling line properties

x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y, linewidth=5.0)
plt.show()
```

```python
# ploting two different lines into one graph

x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [1,4,6,8]

lines = plt.plot(x1,y1, x2,y2)

plt.setp(lines[0], color='r', linewidth=2.0)
plt.setp(lines[1],color='b',linewidth=2.0)

plt.show()
```

```python
def f(t):
    return  np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)

plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), '-b')

plt.subplot(212)
plt.grid()
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
